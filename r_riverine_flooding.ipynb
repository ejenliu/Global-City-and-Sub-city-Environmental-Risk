```python
# Set environment settings
arcpy.env.overwriteOutput = True
# Check in the Spatial Analyst extension license
arcpy.CheckInExtension("Spatial")

def clip_raster_to_polygons(raster_path, shapefile_path, output_path):
    # Use ExtractByMask to clip the raster with the polygon shapefile
    #out_extract_by_mask = arcpy.sa.ExtractByMask(raster_path, shapefile_path)
    # Save the output
    #out_extract_by_mask.save(output_path)
    try:
        # Use ExtractByMask to clip the raster with the polygon layer
        out_extract_by_mask = arcpy.sa.ExtractByMask( raster_path,
        in_mask_data=shapefile_path,
        extraction_area="INSIDE" )
        out_extract_by_mask.save(output_path)
        
        # Check if the output is empty
        if out_extract_by_mask is None:
            print(f"No overlap between {os.path.basename(raster_path)} and cities. Skipping...")
        else:
            # Save the output
            out_extract_by_mask.save(output_path)
            print(f"Saved clipped raster to {output_path}")
    except Exception as e:
        print(f"Error processing {os.path.basename(raster_path)}: {e}")

def process_rasters(raster_paths, shapefile_path, output_dir):
    if not os.path.exists(output_dir):
        os.makedirs(output_dir)

    for raster_path in raster_paths:
        raster_name = os.path.basename(raster_path)
        output_path = os.path.join(output_dir, raster_name)

        clip_raster_to_polygons(raster_path, shapefile_path, output_path)
        print(f"loop done for {raster_name}")


shapefile_path = r"G:\CASA\ejen\cities_dataset_progress_7_11.gpkg\main.cities_dataset_progress_7_11"
output_dir = r"D:\JFM_tests\CASA\riverine_flood_seasonality_cities_tiles"

# Check out the Spatial Analyst extension license
arcpy.CheckOutExtension("Spatial")

process_rasters(seasonality_tifs, shapefile_path, output_dir)

```

```python
arcpy.env.workspace = output_dir
seasonality_tifs = arcpy.ListRasters("*.tif")
seasonality_tifs
```

```python
binary_folder = r"D:\JFM_tests\CASA\riverine_binary_flood_seasonality_tiles"
# Process each raster
for raster in seasonality_tifs:
    # Define the output raster name
    out_raster_name = f"binary_{raster}"
    out_raster_path =  os.path.join(binary_folder, out_raster_name)

    # Apply the conditional statement to create binary output
    binary_raster = arcpy.sa.Con((arcpy.sa.Raster(raster) >= 1) & (arcpy.sa.Raster(raster) <= 11), 1, None)
    
    # Save the output raster
    # binary_raster.save(out_raster_path)
    arcpy.env.compression = "LZW"# Using LZW compression to reduce file size

    # Set the output raster to 8-bit unsigned integer
    arcpy.management.CopyRaster(binary_raster, out_raster_path, 
                                pixel_type="1_BIT", 
                                format="TIFF",
                                nodata_value="0"
                                )
```

```python
arcpy.env.workspace = binary_folder
seasonality_tifs = arcpy.ListRasters("*.tif")
seasonality_tifs
```

```python
seasonality_string = ";".join(seasonality_tifs)
arcpy.management.MosaicToNewRaster(
    input_rasters=seasonality_string,
    output_location=r"D:\JFM_tests\CASA",
    raster_dataset_name_with_extension="r_riverine_flooding_seasonality_merged.tif",
    coordinate_system_for_the_raster='GEOGCS["GCS_WGS_1984",DATUM["D_WGS_1984",SPHEROID["WGS_1984",6378137.0,298.257223563]],PRIMEM["Greenwich",0.0],UNIT["Degree",0.0174532925199433]]',
    pixel_type="1_BIT",
    cellsize=None,
    number_of_bands=1,
    mosaic_method="MAXIMUM",
    mosaic_colormap_mode="FIRST"
)
```
