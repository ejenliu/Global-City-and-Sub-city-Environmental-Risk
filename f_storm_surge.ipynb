```python
global_dem = r"G:\CASA\ejen\Future Storm Surge\dem_files\global cities lecz dem\global_cities_lecz_dem.tif"
```

```python
#Interpolate RP100 storm surge values across LECZ extent
coast_variables_shape = r"G:\CASA\data\COAST-RP\points\COAST_RP_allvar.shp"
lecz_rp100 = r"G:\CASA\output_shapefiles\COAST_st_rp_0100.tif"
outputCoordinateSystem='GEOGCS["GCS_WGS_1984",DATUM["D_WGS_1984",SPHEROID["WGS_1984",6378137.0,298.257223563]],PRIMEM["Greenwich",0.0],UNIT["Degree",0.0174532925199433]]' 
snapRaster=global_dem
extent=global_dem
cellSize=global_dem
mask=global_dem
arcpy.sa.IDW(
    in_features=coast_variables_shape,
    z_field="st_rp_0100",
    out_raster= lecz_rp100,
    cell_size= global_dem,
    power=2,
    search_neighborhood="NBRTYPE=Standard S_MAJOR=99.342614921292 S_MINOR=99.342614921292 ANGLE=0 NBR_MAX=15 NBR_MIN=10 SECTOR_TYPE=ONE_SECTOR",
    weight_field=None
)
```

```python
interpolated_storm_surge = r"G:\CASA\output_rasters\COAST StormTide Return Periods\COAST_st_rp_0100.tif"
```

```python
#Difference storm surge and dem values to identify flooded areas
#Positive vaues indicate flood and negative values indicate no flood
import arcpy
from arcpy.ia import *
outMinus = Minus(interpolated_storm_surge, global_dem)
difference_raster = r"G:\CASA\output_rasters"
arcpy.env.workspace = difference_raster
outMinus.save("difference_storm_surge_rp100.tif")
```

```python
#Pull in population tif for 2025
pop_25 = r"G:\CASA\ejen\City Boundaries and Characteristics\GHS_POP_E2025_GLOBE_R2023A_4326_3ss_V1_0-003.tif"
difference_storm_surge = r"G:\CASA\output_rasters\difference_storm_surge_rp100.tif"
```

```python
import arcpy
from arcpy.sa import *
#Create binary mask for areas where elevation is lower than interpolated storm surge values
mask_raster = Con(Raster(difference_storm_surge) > 0, 1)
```

```python
#Mask population raster using the binary flood layer to keep only population values where flood is expected 
masked_population = Raster(pop_25) * mask_raster
output_masked_population = r"G:\CASA\output_rasters\future_surge_masked_population.tif"
masked_population.save(output_masked_population)
```

```python
#Extract values only where cities are present
lecz_lt10_cities = r"G:\CASA\output_rasters\merit_leczs_lt10_resample_cities.tif"
pop25_lecz = r"D:\JFM_tests\CASA\GHSPop_2025_maskedCitiesLECZ.tif"
arcpy.env.cellSize=pop_25
out_raster = arcpy.sa.ExtractByMask(
    in_raster=pop_25,
    in_mask_data=lecz_lt10_cities,
    extraction_area="INSIDE",
    analysis_extent=lecz_lt10_cities
)
out_raster.save(pop25_lecz)
```
